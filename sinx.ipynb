```python
import numpy as np 
import pandas as pd 
import sklearn as sl 
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
x = np.linspace(-10, 10, 1000)
y = np.sin(x)

df = pd.DataFrame(
    {
        'x': x,
        'y': y
    }
)
```

```python
plt.figure(figsize=(16, 5))
plt.plot(df['x'], df['y'])
```
